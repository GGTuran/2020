```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import  accuracy_score
```

```python
df = pd.read_csv('/content/Copy of sonar data.csv', header=None)
```

```python
X = df.drop(columns=60, axis=1)
Y = df[60]
```

```python
X
```

```python
Y
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.1, stratify=Y, random_state=1)
```

```python
X_train
```

```python
Y_train
```

```python
X_test
```

```python
Y_test
```

```python
model = LogisticRegression()
```

```python
model.fit(X_train, Y_train)
```

```python
# accuracy on training data
X_train_prediction = model.predict(X_train)
score = accuracy_score(X_train_prediction, Y_train)
print("Accuracy score for training data", score)
```

```python
# accuracy on test data
X_test_prediction = model.predict(X_test)
score = accuracy_score(X_test_prediction, Y_test)
print("Accuracy score for test data", score)
```
